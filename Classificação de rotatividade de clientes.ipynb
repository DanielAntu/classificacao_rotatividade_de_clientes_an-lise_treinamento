```python
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
def mostrar_acuracia(modelo, X_train, X_test, y_train, y_test):
  y_pred = modelo.predict(X_test)
  y_pred_train = modelo.predict(X_train)
  print('='*25)
  print('Teste')
  print('='*25)
  print(accuracy_score(y_test, y_pred))
  print(confusion_matrix(y_test, y_pred))
  print('='*25)
  print('Treino')
  print('='*25)
  print(accuracy_score(y_train, y_pred_train))
  print(confusion_matrix(y_train, y_pred_train))

def sgd(X_train, X_test, y_train, y_test):
  sgd = SGDClassifier()
  sgd.fit(X_train, y_train)
  mostrar_acuracia(sgd, X_train, X_test, y_train, y_test)

def svc(X_train, X_test, y_train, y_test):
  svc = SVC()
  svc.fit(X_train, y_train)
  mostrar_acuracia(svc, X_train, X_test, y_train, y_test)

def knn(X_train, X_test, y_train, y_test, n_neighbors=5):
  knn = KNeighborsClassifier(n_neighbors=n_neighbors)
  knn.fit(X_train, y_train)
  mostrar_acuracia(knn, X_train, X_test, y_train, y_test)

def gausiano(X_train, X_test, y_train, y_test):
  gausiano = GaussianProcessClassifier()
  gausiano.fit(X_train, y_train)
  mostrar_acuracia(gausiano, X_train, X_test, y_train, y_test)

def arvore(X_train, X_test, y_train, y_test, criterion='gini', splitter='best', max_depth=None):
  arvore = DecisionTreeClassifier(criterion=criterion, splitter=splitter, max_depth=max_depth, random_state=42)
  arvore.fit(X_train, y_train)
  mostrar_acuracia(arvore, X_train, X_test, y_train, y_test)

def floresta(X_train, X_test, y_train, y_test, n_estimators=100, criterion='gini', max_depth=None):
  floresta = RandomForestClassifier(n_estimators=n_estimators, criterion=criterion, max_depth=max_depth, random_state=42)
  floresta.fit(X_train, y_train)
  mostrar_acuracia(floresta, X_train, X_test, y_train, y_test)
```
